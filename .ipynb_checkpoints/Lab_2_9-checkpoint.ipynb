```python
import pandas as pd
import numpy as np

import pymysql
from sqlalchemy import create_engine

from getpass import getpass  # To get the password without showing the input
```

```python
# This just opens a connection
connection_string = 'mysql+pymysql://root:'+password+'@localhost/sakila' # it is easier to specify the database
engine = create_engine(connection_string)
```

1) How many distinct (different) actors' last names are there?

```python
query = '''SELECT COUNT(DISTINCT last_name) 
FROM sakila.actor ''';

print(query)
data = pd.read_sql_query(query, engine)
data.head()
```

2) Add an additional column day_type with values 'weekend' and 'workday' depending on the rental day of the week.


```python
query = '''SELECT 
    *,
    DATE_FORMAT(CONVERT( LEFT(rental_date, 10) , DATE),
            '%%M') AS 'rental_date_month',
    DATE_FORMAT(CONVERT( LEFT(rental_date, 10) , DATE),
            '%%W') AS 'rental_date_weekday',
    CASE
        WHEN
            DATE_FORMAT(CONVERT( LEFT(rental_date, 10) , DATE),
                    '%%W') = 'Saturday'
                OR 'Sunday'
        THEN
            'weekend'
        ELSE 'workday'
    END AS 'rental_date_day_type'
FROM
    rental ''';

print(query)
data = pd.read_sql_query(query, engine)
data.head()
```

3) Get all films with ARMAGEDDON in the title.

```python
query = '''SELECT * 
FROM sakila.film 
WHERE title LIKE "%%ARMAGEDDON%%"''';

print(query)
data = pd.read_sql_query(query, engine)
data.head()
```

4) Get the 10 longest films.

```python
query = '''SELECT * 
FROM film 
ORDER BY length 
DESC LIMIT 10''';

print(query)
data = pd.read_sql_query(query, engine)
data.head()
```

5) How many films include Behind the Scenes content?

```python
query = '''SELECT COUNT(film_id) 
FROM film 
WHERE special_features Like '%%Behind the Scenes%%' ''';

print(query)
data = pd.read_sql_query(query, engine) 
data.head()
```

6) Which kind of movies (rating) have a mean duration of more than two hours?

```python
query = '''SELECT rating, ROUND(AVG(film_id),2) AS 'average_length_of_film' 
FROM film 
GROUP BY rating 
HAVING average_length_of_film>120 ''';
print(query)
data = pd.read_sql_query(query, engine) 
data.head()
```

7) Rank films by length (filter out the rows that have nulls or 0s in length column). In your output, only select the columns title, length, and the rank.

```python
query = '''SELECT title, length, RANK() OVER(ORDER BY length ASC) AS 'rank' 
FROM film 
WHERE length <> ' ' or length <> 0 ''';

print(query)
data = pd.read_sql_query(query, engine) 
data.head(10)
```
